```python
# Merging All the DataFrames
df = pd.merge(df_01, df_02, on='Date', how='outer')
df = pd.merge(df, df_03, on='Date', how='outer')
df = pd.merge(df, df_04, on='Date', how='outer')
df = pd.merge(df, df_05, on='Date', how='outer')
df = pd.merge(df, df_06, on='Date', how='outer')
df = pd.merge(df, df_07, on='Date', how='outer')
df = pd.merge(df, df_08, on='Date', how='outer')
df = pd.merge(df, df_09, on='Date', how='outer')
df = pd.merge(df, df_10, on='Date', how='outer')
df = pd.merge(df, df_11, on='Date', how='outer')
df = pd.merge(df, df_12, on='Date', how='outer')
df = pd.merge(df, df_13, on='Date', how='outer')

df.head().append(df.tail())
```
